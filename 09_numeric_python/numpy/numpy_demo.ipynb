## Basic Operations

```python
array = np.array([[1,2,3,4],[5,6,7,8]])

print(array)
print()

print("array + 1")
print(array + 1)
print()

print("array * 2")
print(array * 2)
print()

print("array / 2")
print(array / 2)
```

```python
print("max: ", np.max(array))
print("min: ", np.min(array))
print("mean: ", np.mean(array))
print("sin: ", np.sin(array))
```

```python
print("shape before: ", array.shape)
print("mean axis=1", np.mean(array,axis=1))
print("mean axis=0", np.mean(array,axis=0))
```

## Indexing

```python
array = np.array([1,2,3,4,5])
print(f"array: {array}")

print("Select a single element: ")
print(array[1])

print("Select a range:")
print(array[1:4])
```

```python
print("Select with open range")
print(array[2:])
print(array[:2])
print(array[:])
```

```python
array = np.array([
    [1,2,3,4],
    [5,6,7,8]
])
# shape: (2,4)

# Select numbers [2,3,4]
print(array[0,1:])
```

```python
# Select numbers [2,6]
print(array[:,1])
```

```python
# Select numbers [[2,3],[6,7]]
print(array[:,1:3])
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = np.array(Image.open("./image.png"))
img = img[:,:,:3]
print("shape: ", img.shape)
print("dtype: ", img.dtype)
print("range: ", img.min(),img.max())

print("normalized: ")
img = img.astype(float)
img /= 255
print("shape: ", img.shape)
print("dtype: ", img.dtype)
print("range: ", img.min(),img.max())

plt.imshow(img)
```

```python
plt.imshow(img[100:300,100:300])
```

```python
copy = img.copy()
copy[100:300,100:300] *= 0.5
plt.imshow(copy)
```

```python
copy = img.copy()
copy[...,0] = 0
plt.imshow(copy)
```

```python
copy = img.copy()
red_channel = copy[:,:,0]
high_red = red_channel > 0.8
print(high_red.shape)
print(high_red.dtype)
plt.imshow(high_red)
```

```python
print(high_red)
```

```python
copy = img.copy()
print(np.mean(copy[high_red,]))
copy[high_red,] = 0.0
print(np.mean(copy[high_red,]))
plt.imshow(copy)
```
